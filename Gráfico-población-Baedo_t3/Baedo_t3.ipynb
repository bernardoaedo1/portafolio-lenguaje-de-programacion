# Población de China por región

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

class China:
    def __init__(self):
        self.Este = {"Shandong": 101.53, "Anhui": 61.03, "Jiangsu": 84.75, "Jiangxi":45.20, "Zhejiang": 54.43, 
                     "Fujian": 41.54, "Taiwan": 23.89, "Shangai": 24.87}
        self.noreste = {"Heilongjiang": 31.85, "Jilin": 24.07, "Liaoning": 43.90, "Beijing": 21.89, "Tianjin": 15.20}
        self.noroeste = { "Qinhai": 5.39, "Gansu": 25.02, "Shaanxi": 37.33, "Xinjiang": 21.81, "Ningxia": 5.88}
        self.norte = {"Hebei": 74.61, "Shanxi": 34.91, "Inner Mongolia": 24.05}
        self.surcentral = {"Henan": 99.36, "Hubei": 58.50, "Hunan": 66.44, "Hainan": 10.08, "Guangdong": 126.01,
                           "Guangxi": 50.13, "Hong Kong": 7.40, "Macao": 0.68}
        self.suroeste = {"Sichuan": 83.67, "Yunnan": 48.00, "Guizhou": 38.56, "Chongqing": 33.97, "Tibet": 3.18}
        
    def poblacion_regional(self):
        regiones = {
            "Este": sum(self.Este.values()),
            "Noreste": sum(self.noreste.values()),
            "Noroeste": sum(self.noroeste.values()),
            "Norte": sum(self.norte.values()),
            "Surcentral": sum(self.surcentral.values()),
            "Suroeste": sum(self.suroeste.values())
        }
        return regiones
    def Lista_regiones(self):
        nombres = list(regiones.keys())
        return  nombres
    
    def Lista_valores(self):
        valores = list(regiones.values())
        return valores

fig, ax = plt.subplots()
china = China()
regiones = china.poblacion_regional()
nombres = china.Lista_regiones()
valores = china.Lista_valores()
ax.bar(nombres, valores, color=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'])
ax.set_title('Población por Región en China (millones)')
ax.set_ylabel('Población (millones)')
ax.yaxis.set_major_locator(mticker.MultipleLocator(50))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Clase que contiene información de China

```python
class China:
    def __init__(self):
        self.Este = {"Shandong": 101.53, "Anhui": 61.03, "Jiangsu": 84.75, "Jiangxi":45.20, "Zhejiang": 54.43, 
                     "Fujian": 41.54, "Taiwan": 23.89, "Shangai": 24.87}
        self.noreste = {"Heilongjiang": 31.85, "Jilin": 24.07, "Liaoning": 43.90, "Beijing": 21.89, "Tianjin": 15.20}
        self.noroeste = { "Qinhai": 5.39, "Gansu": 25.02, "Shaanxi": 37.33, "Xinjiang": 21.81, "Ningxia": 5.88}
        self.norte = {"Hebei": 74.61, "Shanxi": 34.91, "Inner Mongolia": 24.05}
        self.surcentral = {"Henan": 99.36, "Hubei": 58.50, "Hunan": 66.44, "Hainan": 10.08, "Guangdong": 126.01,
                           "Guangxi": 50.13, "Hong Kong": 7.40, "Macao": 0.68}
        self.suroeste = {"Sichuan": 83.67, "Yunnan": 48.00, "Guizhou": 38.56, "Chongqing": 33.97, "Tibet": 3.18}
```

La información contenida aquí corresponde a la población de china por cada provincia repartidas en diccionarios según la región geográfica a la que pertenezcan.

El país se divide en provincias, regiones autonomas y regiones especiales pero por motivos de conveniencia las regiones autonomas (Xinjiang por ejemplo) y regiones especiales (Hong Kong) serán incluidas dentro de los diccionarios igualmente.

### Método población regional

```python
def poblacion_regional(self):
    regiones = {
        "Este": sum(self.Este.values()),
        "Noreste": sum(self.noreste.values()),
        "Noroeste": sum(self.noroeste.values()),
        "Norte": sum(self.norte.values()),
        "Surcentral": sum(self.surcentral.values()),
        "Suroeste": sum(self.suroeste.values())
    }
    return regiones
```

Este metodo entrega un diccionario cuyas claves corresponden a las regiones geográficas y su valores son la suma de los valores de los diccionarios anteriores, así se tienen los valores totales. El motivo de esto es que la fuente de información no entregaba valores totales de población por región geográfica.

### Método lista de regiones

```python
def Lista_regiones(self):
    nombres = list(regiones.keys())
    return  nombres
```

Este método genera una lista de nombre usando las claves del diccionario generado por el método anterior.

> **[AUDITORÍA DE CÓDIGO] Error Detectado: Ámbito de Variables (Scope)**
>
> **Descripción del error:**
> En los métodos `Lista_regiones` y `Lista_valores`, el código hace referencia directa a la variable `regiones` (`list(regiones.keys())`).
>
> **Análisis:**
> Esta variable `regiones` no pertenece a la clase (no usa `self`) ni se recibe como argumento. El código funciona únicamente porque existe una variable global con el mismo nombre definida en el bloque de ejecución principal.
>
> **Corrección Teórica:**
> Para cumplir con el encapsulamiento de la POO, la clase debería acceder a sus propios datos llamando a su método interno o guardando el resultado en un atributo:
> * *Incorrecto:* `nombres = list(regiones.keys())`
> * *Correcto:* `nombres = list(self.poblacion_regional().keys())`

### Método lista de valores

```python
def Lista_valores(self):
    valores = list(regiones.values())
    return valores
```

Al igual que el metodo anterios este genera una lista pero usando los valores del diccionario generado por el método de población regional.

## Creación del gráfico de barras

```python
fig, ax = plt.subplots()
china = China()
regiones = china.poblacion_regional()
nombres = china.Lista_regiones()
valores = china.Lista_valores()
ax.bar(nombres, valores, color=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'])
ax.set_title('Población por Región en China (millones)')
ax.set_ylabel('Población (millones)')
ax.yaxis.set_major_locator(mticker.MultipleLocator(50))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
